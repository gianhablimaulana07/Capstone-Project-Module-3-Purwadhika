## 1. Business Understanding

### Context

Seiring dengan pertumbuhan populasi dan urbanisasi di Arab Saudi, kebutuhan akan mobil sebagai alat transportasi semakin meningkat. Mobil tidak hanya berfungsi sebagai sarana mobilisasi individu, tetapi juga sebagai alat untuk pengiriman barang. Selain itu, rendahnya biaya bahan bakar dan iklim yang kurang mendukung penggunaan sepeda motor menjadikan mobil sebagai pilihan utama bagi masyarakat Arab Saudi.

Dalam beberapa tahun terakhir, pasar mobil bekas di Arab Saudi telah berkembang pesat akibat meningkatnya permintaan akan kendaraan yang terjangkau. Calon pembeli mobil di Arab Saudi cenderung mencari penawaran terbaik dengan mempertimbangkan berbagai faktor seperti merek, model, tahun pembuatan, kondisi, dan harga. Namun, karena adanya keragaman dalam spesifikasi dan kondisi mobil, calon pembeli sering kali kesulitan untuk menilai harga yang adil untuk kendaraan yang mereka minati. Oleh karena itu, perusahaan yang beroperasi di sektor otomotif, baik itu platform jual beli mobil atau dealer, perlu memahami faktor-faktor yang mempengaruhi harga mobil bekas.

### Problem Statement

Dalam pasar mobil bekas di Arab Saudi, salah satu tantangan utama adalah menentukan harga jual yang kompetitif dan akurat bagi kendaraan yang akan dipasarkan. Data yang tidak konsisten dan beragam tentang harga mobil bekas mengakibatkan kesulitan bagi pembeli dan penjual dalam mengambil keputusan yang tepat. Konsumen sering kali bingung apakah harga yang mereka tawarkan atau diterima sesuai dengan tren pasar saat ini. Selain itu, penjual membutuhkan panduan untuk menetapkan harga yang wajar agar bisa bersaing di pasar, tanpa merugikan diri mereka sendiri.

Oleh karena itu, perlu dikembangkan model machine learning untuk memprediksi harga mobil bekas yang didasarkan pada berbagai fitur seperti merek, model, tahun produksi, tipe mesin, kilometer yang telah ditempuh, dan kondisi mobil. Dengan pendekatan ini, diharapkan model bisa memberikan estimasi yang akurat mengenai harga, sehingga pembeli dapat membuat keputusan yang lebih informasi dan penjual dapat menetapkan harga jual yang sesuai dengan kondisi pasar.

## 4. Exploratory Data Analysis (EDA)

*Exploratory Data Analysis* (EDA) digunakan untuk memahami struktur, hubungan, dan pola dalam data yang diberikan. Tujuan dari EDA adalah untuk mendapatkan wawasan yang mendalam tentang data sebelum menjalankan analisis statistik yang lebih lanjut.

Kolom *Price* adalah variabel target dan kolom-kolom lainnya adalah variabel independen. Variabel independen kembali dibagi menjadi Numerical Variabel dan Categorical Variabel. Berikut rinciannya:

- **Numerical Variabels** : Year, Engine_Size, Mileage, Price
- **Categorical Variabels** : Type, Region, Make, Gear_Type, Origin, Options, Negotiable

### Numerical Variabel

Untuk variabel numerik, kita dapat mengetahui korelasi dengan menggunakan Pair Plot. Pair Plot akan menggambar pasangan dari beberapa variabel numerik dalam daftar sehingga kita dapat melihat korelasi antara setiap variabel numerik.

Dengan demikian, pair plot membantu kita untuk memahami korelasi antara variabel-variabel numerik dalam dataset kita dan dapat memberikan wawasan yang lebih baik tentang bagaimana variabel-variabel ini mempengaruhi harga mobil bekas.

```python
sns.pairplot(df[['Year','Engine_Size','Mileage','Price']], aspect=1.2)
```

**Insight:**

1. Terdapat korelasi positif yang kuat antara **Engine_Size** dan **Price**.
Artinya, semakin besar ukuran mesin, semakin tinggi harga mobil bekas tersebut.
2. Terdapat korelasi negatif yang lemah antara **Year** dan **Price**. Artinya, semakin tua tahun pembuatan mobil, semakin rendah harga mobil bekas tersebut.
3. Terdapat korelasi negatif yang lemah antara **Mileage** dan **Price**. Artinya, semakin tinggi jarak tempuh mobil, semakin rendah harga mobil bekas tersebut.


### Categorical Variabel

Untuk Variabel Kategorikal, kita akan menganalisis nilai pada setiap kategori untuk melakukan eksplorasi dan merangkum kategori itu sendiri. Kita akan melakukan peringkat dan menghitung nilai pada setiap kategori, sehingga kita dapat melihat perusahaan mobil bekas mana yang paling banyak, jenis mobil bekas apa yang paling banyak, wilayah mana yang paling tinggi penggunaan mobil bekas, dan sebagainya.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Type
sns.countplot(data=df, x='Type', order=df['Type'].value_counts().iloc[:10].index, palette='tab10', ax=axes[0])
axes[0].set_title('Top 10 Tipe Mobil Bekas Terbanyak')
axes[0].set_xlabel('Type')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)
plt.setp(axes[0].get_xticklabels(), ha="right")

# Make
sns.countplot(data=df, x='Make', order=df['Make'].value_counts().iloc[:10].index, palette='tab10', ax=axes[1])
axes[1].set_title('Top 10 Perusahaan Mobil Bekas Terbanyak')
axes[1].set_xlabel('Make')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)
plt.setp(axes[1].get_xticklabels(), ha="right")

# Adjust layout
plt.tight_layout()
plt.show()
```

**Insight:**

Grafik di atas menunjukkan bahwa Toyota adalah perusahaan penghasil mobil bekas yang paling banyak daripada perusahaan lainnya. Hal ini terkait karena dalam Top 10 Tipe Mobil Bekas diisi dengan Land Cruiser, Campry, Hilux & Corolla. Selanjutnya, Hyundai membuat mobil bekas yang kedua paling banyak diproduksi dengan Accent, Sonata, dan Elantra mereka. Dengan kata lain, regresi linier digunakan untuk menentukan sejauh mana satu atau lebih variabel dapat memprediksi nilai variabel dependen.

**Ada dua jenis utama Linear Regression:**

a. Linear Regression Sederhana:

Linear Regression sederhana adalah pendekatan untuk memprediksi respons menggunakan satu fitur. Ini adalah salah satu model machine learning paling dasar yang harus diketahui. Linear Regression Berganda adalah metode statistik yang digunakan untuk memodelkan hubungan linier antara variabel dependen dan beberapa variabel independen. Tujuan Linear Regression Berganda adalah untuk memodelkan hubungan linier antara variabel independen dan variabel respons (dependen).

Metode ini termasuk dalam kategori algoritma non-parametrik yang bekerja dengan mengukur jarak (seperti Euclidean distance) antara titik data yang ada dan data baru yang akan diprediksi. KNN Regressor sering digunakan dalam berbagai aplikasi, termasuk analisis data, prediksi, dan kontrol sistem.

**Cara Kerja KNN Regressor**

a. Pemilihan Parameter K: Tentukan jumlah tetangga terdekat yang akan dipertimbangkan.

b. Pengukuran Jarak: Hitung jarak antara data titik yang baru dan data titik yang ada menggunakan metrik yang sesuai (misalnya, Euclidean).

c. Pengambilan K Tetangga Terdekat: Temukan k data terdekat berdasarkan jarak yang telah dihitung.

d. Perhitungan Rata-rata: Untuk KNN Regressor, prediksi output untuk data baru adalah rata-rata dari output k tetangga terdekat.

**Definisi Decision Tree Regression**

Decision Tree Regression adalah metode pemodelan statistik yang digunakan untuk memprediksi nilai numerik berdasarkan fitur-fitur input. Pada dasarnya, decision tree adalah struktur pohon yang terdiri dari simpul (nodes) dan cabang (branches) yang digunakan untuk membuat keputusan berdasarkan serangkaian pertanyaan. Dalam konteks regresi, tujuan utama adalah mempartisi ruang input ke dalam segmen-segmen yang homogen sehingga kita dapat memprediksi nilai target untuk setiap segmen dengan akurasi yang tinggi.

**Cara Kerja Decision Tree Regression**

a. **Pemilihan Fitur**:

Decision tree mencari fitur input yang terbaik untuk membagi dataset. Proses ini berdasarkan pada seberapa baik fitur tersebut dapat memisahkan data ke dalam kelompok yang berbeda. Kriteria pemisahan ini bisa menggunakan metrik seperti Mean Squared Error (MSE).

b. **Pembentukan Simpul**:

Setiap simpul dalam pohon merepresentasikan keputusan berdasarkan fitur input. Misalnya, jika kita menggunakan fitur X1, simpul mungkin berisi pertanyaan seperti "Apakah X1 > 5?". Berdasarkan jawaban (ya atau tidak), kita akan menuju ke simpul berikutnya.

c. **Pemisahan Rekursif**:

Proses ini dilakukan secara rekursif untuk setiap subset dari data hingga mencapai kondisi tertentu, seperti kedalaman maksimum pohon atau jumlah minimum data pada simpul akhir.

d. **Leaf Nodes**:

Ketika proses pemisahan selesai, simpul akhir yang tidak dapat dibagi lagi disebut sebagai leaf node. Nilai target untuk leaf node biasanya merupakan rata-rata dari nilai target yang berada di dalam node tersebut.

e. **Pruning (Pemangkasan)**:

Proses ini membantu mengurangi kompleksitas model dengan menghapus cabang-cabang yang tidak memberikan peningkatan signifikan terhadap akurasi. Ini menghindari overfitting.

d. **Seleksi Fitur:**

Dengan mengeluarkan koefisien yang tidak signifikan (menjadi nol), Lasso membantu dalam seleksi fitur, yang menghasilkan model yang lebih interpretatif dan efisien.

MAE memberikan bobot yang sama untuk semua kesalahan, sehingga akan memberikan informasi yang jelas mengenai kesalahan yang terjadi, tanpa mengganggu karena besarnya kesalahan (seperti halnya pada MSE).

3. **MAPE (Mean Absolute Percentage Error)** [(Source)](https://www.sciencedirect.com/science/article/abs/pii/S0169207006000239)

**Definisi MAPE:**

MAPE (Mean Absolute Percentage Error) adalah sebuah metrik yang digunakan untuk mengukur akurasi dari model peramalan. MAPE dihitung dengan mengambil rata-rata dari selisih absolut antara nilai yang diprediksi dan nilai aktual, dibagi oleh nilai aktual, dan dinyatakan dalam persentase.

Rumus MAPE adalah sebagai berikut:

image.png

**Interpretasi MAPE**:

a. Kualitas Peramalan:

MAPE memberikan gambaran tentang seberapa baik model dalam memprediksi nilai. Semakin kecil nilai MAPE, semakin baik akurasi model.

b. Persentase Kesalahan:

Karena MAPE dinyatakan dalam persentase, mudah untuk memahami ukuran kesalahan relatif terhadap nilai aktual. Misalnya, MAPE sebesar 10% menunjukkan bahwa rata-rata kesalahan dari semua prediksi adalah 10% dari nilai aktual.

c. Keterbatasan:

MAPE bisa menjadi tidak terdefinisi jika ada nilai ( A_i ) yang sama dengan nol serta dapat memberikan bobot yang lebih besar pada kesalahan untuk nilai kecil. Oleh karena itu, perlu diinterpretasikan dengan hati-hati, terutama dalam dataset yang memiliki nilai aktual yang sangat bervariasi.

4. **R (R-squared)** [(Source)](https://www.statlearning.com/)

**Definisi R (R-squared):**

R (R-squared) adalah sebuah ukuran statistik yang digunakan untuk menilai seberapa baik model regresi dapat menjelaskan variasi dalam data. R menunjukkan proporsi varians dalam variabel dependen yang dapat dijelaskan oleh variabel independen dalam model. Nilai R berkisar antara 0 dan 1, di mana:

- R = 0: Model tidak menjelaskan variasi apapun dalam data; semua variabilitas dalam hasil adalah residual.
- R = 1: Model menjelaskan semua variasi dalam data; semua data sesuai dengan garis regresi.
- R berada di antara 0 dan 1: Model menjelaskan sebagian dari variasi dalam data.

image.png

**Interpretasi R:**

- R yang lebih tinggi menunjukkan bahwa model memiliki kekuatan prediktif yang lebih baik. Misalnya, nilai R sebesar 0,8 berarti 80% dari variasi dalam variabel dependen dapat dijelaskan oleh variabel independen dalam model, sementara 20% sisanya disebabkan oleh faktor lain atau variabilitas acak.
- Penting untuk diingat bahwa R tidak dapat digunakan sebagai satu-satunya ukuran untuk memvalidasi model. Meskipun R tinggi, model bisa saja overfitting, yaitu terlalu cocok dengan data pelatihan tetapi berkinerja buruk pada data baru.

## 7. Hyperparameter Tuning

Pada tahap ini, kita akan melakukan hyperparameter tuning pada model CatBoostRegressor dengan harapan dapat meningkatkan performa model.

### Hyperparameter tuning using RandomizedSearchCV

```python
# Hyperparameter tuning using RandomizedSearchCV
param_dist = {
    'regressor__iterations': [500, 1000, 1500],
    'regressor__learning_rate': [0.01, 0.05, 0.1],
    'regressor__depth': [4, 6, 8],
    'regressor__l2_leaf_reg': [1, 3, 5]
}

# Wrap the CatBoostRegressor in a pipeline with transformers
pipeline_catboost = Pipeline([
    ('transformer', transformer),
    ('regressor', CatBoostRegressor(verbose=0))
])

random_search = RandomizedSearchCV(
    pipeline_catboost,
    param_distributions=param_dist,
    n_iter=10,
    cv=5,
    scoring='neg_root_mean_squared_error',
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)
```

Hasil dari penelusuran parameter menggunakan RandomizedSearchCV menunjukkan bahwa parameter terbaik untuk model CatBoostRegressor adalah sebagai berikut:

**a. 'regressor__learning_rate': 0.05**

Parameter ini menentukan tingkat pembelajaran atau seberapa cepat model belajar dari data. Dalam kasus ini, tingkat pembelajaran adalah 0.05.

**b. 'regressor__l2_leaf_reg': 5**

Parameter ini merupakan regulasi L2 yang digunakan dalam model untuk mencegah overfitting. Nilai 5 menunjukkan kekuatan regulasi yang diterapkan pada model.

**c. 'regressor__iterations': 1000**

Parameter ini menentukan jumlah iterasi atau langkah yang akan dilakukan dalam proses gradient boosting. Model akan mempelajari pola data dalam sebanyak 1000 iterasi.

**d. 'regressor__depth': 8**

Parameter ini mengontrol kedalaman maksimum dari pohon yang digunakan dalam model. Dalam kasus ini, kedalaman maksimum pohon adalah 8.

### Predict to Test Set with the Tuned Model
